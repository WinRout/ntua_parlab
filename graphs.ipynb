# Graphs for parlab

## kmeans

```python
import matplotlib.pyplot as plt
import numpy as np

file = open('a2/kmeans/kmeans_seq.out')
content = file.readlines()
seq_conf1 = float(content[34][32:39])

file = open('a2/kmeans/kmeans_seq_conf2.out')
content = file.readlines()
seq_conf2 = float(content[22][32:39])

naive = [seq_conf1]
naive_affinity = [seq_conf1]
reduction_conf1 = [seq_conf1]
reduction_conf1_fs = [seq_conf1]
reduction_conf2 = [seq_conf2]
reduction_conf2_fs = [seq_conf2]

file = open('a2/kmeans/kmeans_omp_naive.out')
content = file.readlines()
for i in range (0,7):
    naive.append(float(content[i*53+34][32:39]))

file = open('a2/kmeans/kmeans_omp_naive_affinity.out')
content = file.readlines()
for i in range (0,7):
    naive_affinity.append(float(content[i*53+34][32:39]))

file = open('a2/kmeans/kmeans_omp_reduction_conf1.out')
content = file.readlines()
for i in range (0,7):
    reduction_conf1.append(float(content[i*53+34][32:39]))

file = open('a2/kmeans/kmeans_omp_reduction_conf1_fs.out')
content = file.readlines()
for i in range (0,7):
    reduction_conf1_fs.append(float(content[i*53+34][32:39]))

file = open('a2/kmeans/kmeans_omp_reduction_conf2.out')
content = file.readlines()
for i in range (0,7):
    reduction_conf2.append(float(content[i*29+22][32:39]))

file = open('a2/kmeans/kmeans_omp_reduction_conf2_fs.out')
content = file.readlines()
for i in range (0,7):
    reduction_conf2_fs.append(float(content[i*29+22][32:39]))

x = ["seq", 1, 2, 4, 8, 16, 32, 64]
plt.rcParams.update({'font.size': 12})

fig, axs = plt.subplots(1,2)
axs[0].plot(x, naive, marker = 'o')
axs[0].set_title('kmeans naive')
axs[0].set(xlabel='seq/threads', ylabel = 'sec')
axs[1].plot(x, naive[0]/np.array(naive), marker = 'o', color='red')
axs[1].set_title('speedup')
axs[1].set(xlabel='seq/threads', ylabel = 'seq_time/time')
fig.set_figwidth(14)
fig.savefig("a2/kmeans/docs/naive.png")

fig, axs = plt.subplots(1,2)
axs[0].plot(x, naive_affinity, marker = 'o')
axs[0].set_title('kmeans naive affinity')
axs[0].set(xlabel='seq/threads', ylabel = 'sec')
axs[1].plot(x, naive_affinity[0]/np.array(naive_affinity), marker = 'o', color='red')
axs[1].set_title('speedup')
axs[1].set(xlabel='seq/threads', ylabel = 'seq_time/time')
fig.set_figwidth(14)
fig.savefig("a2/kmeans/docs/naive_affinty.png")

fig, axs = plt.subplots(1,2)
axs[0].plot(x, reduction_conf1, marker = 'o')
axs[0].set_title('kmeans reduction 16x16')
axs[0].set(xlabel='seq/threads', ylabel = 'sec')
axs[1].plot(x, reduction_conf1[0]/np.array(reduction_conf1), marker = 'o', color='red')
axs[1].set_title('speedup')
axs[1].set(xlabel='seq/threads', ylabel = 'seq_time/time')
fig.set_figwidth(14)
fig.savefig("a2/kmeans/docs/reduction_16x16.png")

fig, axs = plt.subplots(1,2)
axs[0].plot(x, reduction_conf1_fs, marker = 'o')
axs[0].set_title('kmeans reduction 16x16 false-sharing optimized')
axs[0].set(xlabel='seq/threads', ylabel = 'sec')
axs[1].plot(x, reduction_conf1_fs[0]/np.array(reduction_conf1_fs), marker = 'o', color='red')
axs[1].set_title('speedup')
axs[1].set(xlabel='seq/threads', ylabel = 'seq_time/time')
fig.set_figwidth(14)
fig.savefig("a2/kmeans/docs/reduction_16x16_fs.png")

fig, axs = plt.subplots(1,2)
axs[0].plot(x, reduction_conf2, marker = 'o')
axs[0].set_title('kmeans reduction 1x4')
axs[0].set(xlabel='seq/threads', ylabel = 'sec')
axs[1].plot(x, reduction_conf2[0]/np.array(reduction_conf2), marker = 'o', color='red')
axs[1].set_title('speedup')
axs[1].set(xlabel='seq/threads', ylabel = 'seq_time/time')
fig.set_figwidth(14)
fig.savefig("a2/kmeans/docs/reduction_1x4.png")

fig, axs = plt.subplots(1,2)
axs[0].plot(x, reduction_conf2_fs, marker = 'o')
axs[0].set_title('kmeans reduction 1x4 false-sharing optimized')
axs[0].set(xlabel='seq/threads', ylabel = 'sec')
axs[1].plot(x, reduction_conf2_fs[0]/np.array(reduction_conf2_fs), marker = 'o', color='red')
axs[1].set_title('speedup')
axs[1].set(xlabel='seq/threads', ylabel = 'seq_time/time')
fig.set_figwidth(14)
fig.savefig("a2/kmeans/docs/reduction_1x4_fs.png")
```
