```python
import matplotlib.pyplot as plt
import numpy as np

file = open('a2/kmeans/kmeans_seq.out')
content = file.readlines()
seq_conf1 = float(content[34][32:39])

file = open('a2/kmeans/kmeans_seq_conf2.out')
content = file.readlines()
seq_conf2 = float(content[22][32:39])

naive = [seq_conf1]
naive_affinity = [seq_conf1]
reduction_conf1 = [seq_conf1]
reduction_conf1_fs = [seq_conf1]
reduction_conf2 = [seq_conf2]
reduction_conf2_fs = [seq_conf2]

file = open('a2/kmeans/kmeans_omp_naive.out')
content = file.readlines()
for i in range (1,7):
    naive.append(float(content[i*53+34][32:39]))

file = open('a2/kmeans/kmeans_omp_naive_affinity.out')
content = file.readlines()
for i in range (1,7):
    naive_affinity.append(float(content[i*53+34][32:39]))

file = open('a2/kmeans/kmeans_omp_reduction_conf1.out')
content = file.readlines()
for i in range (1,7):
    reduction_conf1.append(float(content[i*53+34][32:39]))

file = open('a2/kmeans/kmeans_omp_reduction_fs_conf1.out')
content = file.readlines()
for i in range (1,7):
    reduction_conf1_fs.append(float(content[i*53+34][32:39]))

file = open('a2/kmeans/kmeans_omp_reduction_conf2.out')
content = file.readlines()
for i in range (1,7):
    reduction_conf2.append(float(content[i*29+22][32:39]))

file = open('a2/kmeans/kmeans_omp_reduction_fs_conf2.out')
content = file.readlines()
for i in range (1,7):
    reduction_conf2_fs.append(float(content[i*29+22][32:39]))

def plot(var1, var2, names, name):
    plt.rcParams.update({'font.size': 12})
    barWidth = 0.25
    x1 = np.arange(len(var1))
    x2 = [x + barWidth for x in x1]
    labels = ['1', '2', '4', '8', '16', '32', '64']
    fig, axs = plt.subplots(1,2)
    axs[0].bar(x1, var1, width=barWidth)
    axs[0].bar(x2, var2, width=barWidth)
    axs[0].set_xticks(x1, labels)
    axs[0].set_title('kmeans '+ name)
    axs[0].set(xlabel='seq/threads', ylabel = 'sec')
    axs[0].legend(names)
    axs[0].grid()
    axs[1].bar(x1, var1[0]/np.array(var1), width=barWidth)
    axs[1].bar(x2, var2[0]/np.array(var2), width=barWidth)
    axs[1].set_xticks(x1, labels)
    axs[1].set_title('kmeans ' + name + ' speedup')
    axs[1].set(xlabel='seq/threads', ylabel = 'seq_time/time')
    axs[1].legend(names)
    axs[1].grid()
    fig.set_figwidth(14)
    fig.savefig('docs/kmeans/'+name+'.png')

plot(naive, naive_affinity, ['naive', '+affinity'], 'naive')
plot(reduction_conf1, reduction_conf1_fs, ['16x16', 'fsoptimized'], 'reduction16x16')
plot(reduction_conf2, reduction_conf2_fs, ['1x4', 'fs optimized'], 'reduction1x4')
```

```python
def read_file(path):
    file=open(path)
    content = file.readlines()
    data = []
    skip_line = 1
    for line in content:
        if (skip_line):
            skip_line = 0
            continue
        else:
            skip_line = 1
            element = []
            skip_word = 1
            for word in line.split():
                if(skip_word):
                    skip_word = 0
                    continue
                else:
                    skip_word = 1
                    element.append(word)
            data.append((int(element[0]), element[2], float(element[3])))
    counter = 0
    final = []
    table = []
    for element in data:
        if (counter != 8):
            table.append(element[2])
            counter += 1
        else:
            final.append(table)
            table = [element[2]]
            counter = 1
    final.append(table)
    return(final)

cgl_1024 = read_file('a2.3/conc_ll/output/cgl_1024.out')
cgl_8192 = read_file('a2.3/conc_ll/output/cgl_8192.out')
fgl_1024 = read_file('a2.3/conc_ll/output/fgl_1024.out')
fgl_8192 = read_file('a2.3/conc_ll/output/fgl_8192.out')
lazy_1024 = read_file('a2.3/conc_ll/output/lazy_1024.out')
lazy_8192 = read_file('a2.3/conc_ll/output/lazy_8192.out')
nb_1024 = read_file('a2.3/conc_ll/output/nv_1024.out')
nb_8192 = read_file('a2.3/conc_ll/output/nv_8192.out')
opt_1024 = read_file('a2.3/conc_ll/output/opt_1024.out')
opt_8192 = read_file('a2.3/conc_ll/output/opt_8192.out')
serial_1024 = read_file('a2.3/conc_ll/output/serial_1024.out')
serial_8192 = read_file('a2.3/conc_ll/output/serial_8192.out')

#ploting

def plot_(case, title):
    
    plt.rcParams.update({'font.size': 12})
    barWidth = 0.07
    x1 = np.arange(len(cgl_1024[case]))
    x2 = [x + barWidth + 0.03 for x in x1]
    x3 = [x + barWidth + 0.03 for x in x2]
    x4 = [x + barWidth + 0.03 for x in x3]

    labels = ['1', '2', '4', '8', '16', '32', '64', '128']
    
    plt.bar(x1, serial_1024[0][case]/np.array(cgl_1024[case]), width=barWidth)
    plt.bar(x2, serial_1024[0][case]/np.array(fgl_1024[case]), width=barWidth)
    plt.bar(x3, serial_1024[0][case]/np.array(lazy_1024[case]), width=barWidth)
    plt.bar(x4, serial_1024[0][case]/np.array(nv_1024[case]), width=barWidth)
    plt.bar(x5, serial_1024[0][case]/np.array(opt_1024[case]), width=barWidth)

    names = ['cgl', 'fgl', 'lazy', 'nv', 'opt']

    ax2.set_xticks(x1, labels)
    ax.set_title(title)
    ax2.set_xlabel=('threads') 
    ax2.set_ylabel = ('speedup')
    ax.legend(names)
    ax2.grid()
    ax.grid()
    fig1.savefig(docs/conc_ll/title+'.png')
```
