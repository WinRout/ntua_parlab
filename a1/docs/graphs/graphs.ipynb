```python
import numpy as np
import matplotlib.pyplot as plt

res64   = [0.023136, 0.013838, 0.010257, 0.009107, 0.009404]
res1024 = [10.969285, 5.457227, 2.722889, 1.828968, 1.376893]
res4096 = [175.6115125, 88.201773, 44.463432, 34.752681, 33.832396]
x = np.array([1,2,4,6,8])

y64 = np.array(res64)*1000
y1024 = np.array(res1024)
y4096 = np.array(res4096)

su64 = y64[0]/y64
su1024 = y1024[0]/y1024
su4096 = y4096[0]/y4096

plt.rcParams.update({'font.size': 12})
fig, axs = plt.subplots(2, 3)
axs[0,0].plot(x, y64, marker = 'o')
axs[0,0].set_title('Game of Life 64x64')
axs[0,0].set(xlabel = 'Number of Threads', ylabel = 'Time (ms)')
axs[0,1].plot(x, y1024, marker = 'o', color = 'orange')
axs[0,1].set_title('Game of Life 1024x1024')
axs[0,1].set(xlabel = 'Number of Threads', ylabel = 'Time (s)')
axs[0,2].plot(x, y4096, marker = 'o', color = 'red')
axs[0,2].set_title('Game of Life 4096x4096')
axs[0,2].set(xlabel = 'Number of Threads', ylabel = 'Time (s)')
axs[1,0].plot(x, su64, marker = 'o')
axs[1,0].set(xlabel = 'Number of Threads', ylabel = 'Speedup Ratio')
axs[1,1].plot(x, su1024, marker = 'o', color = 'orange')
axs[1,1].set(xlabel = 'Number of Threads', ylabel = 'Speedup Ratio')
axs[1,2].plot(x, su4096, marker = 'o', color = 'red')
axs[1,2].set(xlabel = 'Number of Threads', ylabel = 'Speedup Ratio')

fig.set_figwidth(12)
fig.set_figheight(10)
fig.savefig("plots.png")
```
